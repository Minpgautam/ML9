```python
#import all libraries
import numpy as np
from sklearn import cluster, datasets
from scipy import misc
from skimage import io
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import scipy.misc
face = scipy.misc.face(gray=True)
X=face.shape
face.max()
```

```python
face.dtype
```

```python
plt.gray()
plt.imshow(face)
plt.show()
```

```python
#basic information about image
reshapeFace = (face / 255.0).reshape(-1,3)

from sklearn.cluster import KMeans
km = KMeans(n_clusters=5)
km.fit(reshapeFace)

print(km.labels_)

print(km.labels_.shape)
```

```python
##create the new image with the clusters=5 and labels
newimg = km.cluster_centers_[km.labels_]

print(newimg.shape)
```

```python
#Reshape the New image with original dimensions
newimg=np.reshape(newimg, (face.shape))

shape=newimg.shape
type=newimg.dtype

print(shape,type)
```

```python
#Plot the compressed image
fig = plt.figure(figsize=(8,8))
ax=fig.add_subplot(1,2,1,xticks=[],yticks=[],title='Acutal\Original Image')
ax.imshow(face)
ax=fig.add_subplot(1,2,2,xticks=[],yticks=[],title='Color Compressed Image using K-Means=5')
ax.imshow(newimg)
plt.show()
```
